```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
import os
os.chdir(r"C:\Training\Academy\Statistics (Python)\Datasets")
```

```python
cars93 = pd.read_csv("Cars93.csv")
print(cars93.info())
```

### Interactive Histogram

```python
def cust_hist(classes):
    sns.histplot(data=cars93, x='Price', bins=classes)
    plt.show()
cust_hist(classes=17)
```

```python
widgets.interact(cust_hist, classes=(4,20,2))
plt.show()
```

```python
def cust_hist(Variable,Classes):
    sns.histplot(data=cars93, x=Variable, bins=Classes)
    plt.show()
cust_hist('EngineSize',10)
```

```python
cars93.columns
```

```python
widgets.interact(cust_hist, Variable=['Price','EngineSize','Wheelbase','Horsepower'], 
                 Classes=(4,20,2))
plt.show()
```

### Interactive Scatter Plot

```python
def cust_scatter(X,Y,Colour):
    sns.scatterplot(data=cars93,x=X,y=Y,hue=Colour)
    plt.show()
cust_scatter(X='EngineSize', Y='Price', Colour='Origin')
```

```python
cars93.select_dtypes(include='number').columns
```

```python
all_numeric = cars93.select_dtypes(include='number').columns
all_categorical = cars93.select_dtypes(exclude='number').columns
```

```python
widgets.interact(cust_scatter, X = all_numeric, Y = all_numeric, Colour = all_categorical)
```

### Interactive Boxplot

```python
def cust_box(X='AirBags', Y='Price'):
    sns.boxplot(x=X,y=Y, data=cars93)
    plt.show()
cust_box(X='Origin', Y='Horsepower')
```

```python
widgets.interact(cust_box, X = all_categorical, Y = all_numeric)
plt.show()
```

### Interactive Bar Chart

```python
def cust_bar(Category):
    cts = cars93[Category].value_counts()
    cts1 = cts.reset_index()
    cts1.columns = [Category, 'Count']
    sns.barplot(data=cts1, x=Category, y='Count')
    plt.show()
```

```python
cust_bar(Category='Origin')
```

```python
widgets.interact(cust_bar, Category=all_categorical)
plt.show()
```

By other option

```python
def cust_bar2(Category):
    cts = cars93[Category].value_counts()
    cts.plot(kind='bar')
    plt.show()
cust_bar2('Origin')
```

```python
widgets.interact(cust_bar2,Category=all_categorical )
```

### Choosing a Chart 

```python
ChartType = "Density"
Variable = 'EngineSize'
if ChartType=="Histogram":
    sns.histplot(data=cars93, x=Variable)
elif ChartType=="Density":
    sns.kdeplot(data=cars93, x=Variable)
elif ChartType == "Box Plot":
    sns.boxplot(y=Variable, data=cars93)
plt.show()
```

```python
def cust_plot(Variable, ChartType, Classes=10):
    if ChartType=="Histogram":
        sns.histplot(data=cars93, x=Variable, bins=Classes)
    elif ChartType=="Density":
        sns.kdeplot(data=cars93, x=Variable)
    elif ChartType == "Box Plot":
        sns.boxplot(y=Variable, data=cars93)
    plt.show()
cust_plot(Variable='Horsepower', ChartType="Histogram", Classes=7)
```

```python
widgets.interact(cust_plot, Variable=all_numeric, ChartType=["Histogram","Density","Box Plot"],
                 Classes=(4,30,2))
plt.show()
```
